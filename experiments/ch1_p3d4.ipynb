```python
import numpy as np
import numpy.linalg as nla
import matplotlib.pyplot as plt
from sympy import symbols, integrate
```

```python
a1, a2, b1, b2, x = sympy.symbols('a1 a2 b1 b2 x')

t1 = integrate((b1+2*b2*(x-3))*(a1+2*a2*(x-3)),(x,1,3))
t2 = -1*integrate((b1*(x-3) + b2*(x-3)**2)*2*x,(x,1,3))
t3 = (0.1*(b1*(x-3) + b2*(x-3)**2)).subs(x,1)
f = t1 + t2 + t3

print(t1)
print(t2)
print(t3)
print(f)
```

```python
a = np.array([[2,-4],[-4,32/3]])*10**4
b = np.array([[-6.4666],[7.6]])
x = nla.solve(a,b)
print(x)
```

```python
print(x)

s_quad = lambda z: (x[0] + 2*x[1]*(z-3))*10**4

z = np.linspace(1,3,50)
y = s_quad(z)

print(y)

fig, ax = plt.subplots(figsize=(10,10))
ax.grid()
ax.plot(z,y)
plt.show()
```

```python
r, R = symbols('r R')
t1 = integrate((b1+2*b2*r)*r*(a1+2*a2*r),(r,0,R))
t2 = -1*integrate((b1*(r-R)+b2*(r**2-R**2))*r,(r,0,R))
t3 = (-1*(b1*(r-R)+b2*(r**2-R**2))*r*(a1+2*a2*r)).subs(r,0)

print(t1)
print(t2)
print(t3)
```
